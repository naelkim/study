```python
x = np.linspace(0, 10, 100)
y = np.arange(0, 10, 0.01)
plt.figure(figsize = (10, 7))
plt.plot(x, np.sin(x))
plt.plot(y, np.cos(y), '--')
plt.show()
```

```python
plt.figure(figsize = (10, 7))
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x), '-')
plt.xlabel('plt.plot(x, np.sin(x), "-")')

plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), '--')
plt.xlabel('plt.plot(x, np.cos(x), "--")')

plt.subplot(2, 2, 3)
plt.plot(x, x)
plt.xlabel('plt.plot(x, x)')

plt.subplot(2, 2, 4)
plt.plot(x, -x)
plt.xlabel('plt.plot(x, -x)')

plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize = (13, 8))

ax[0][0].plot(x, np.sin(x), '-')
ax[0][1].plot(x, np.cos(x), '--')
ax[1][0].plot(x, x)
ax[1][1].plot(x, -x);
```

bar plot


```python
file = os.listdir()[-1]
pop = pd.read_csv(file)
```

```python
plt.figure(figsize = (12, 8))
plt.bar(pop.loc[0:10, 'municipality'], pop.loc[:10, 'pop10'])
plt.show()
```

```python
plt.figure(figsize = (12, 8))
plt.bar(pop.loc[0:10, 'municipality'], pop.loc[:10, 'pop10'] / 1000)
plt.title('Population of major cities(2010)')
plt.xlabel('Municipality')
plt.ylabel('Population(in thousands)')
plt.show()
```

```python
plt.figure(figsize = (12, 8))
plt.barh(pop.loc[0:10, 'municipality'], pop.loc[:10, 'pop10'] / 1000)
plt.title('Population of major cities(2010)')
plt.xlabel('Municipality')
plt.ylabel('Population(in thousands)')
plt.show()
```

```python
import random

number_of_colors = 10 

color = ['#'+''.join([random.choice('0123456789ABCDEF') for j in range(6)]) for i in range(number_of_colors)]
print(color)
```

```python
plt.figure(figsize = (12, 8))
plt.bar(pop.loc[0:10, 'municipality'], pop.loc[:10, 'pop10'] / 1000, color = color)
plt.title('Population of major cities(2010)')
plt.xlabel('Municipality')
plt.ylabel('Population(in thousands)')
plt.show()
```

```python
plt.figure(figsize = (12, 8))
x = np.arange(11)
width = 0.3

plt.bar(x - width, pop.loc[:10, 'pop10'] / 1000, width, label = '2010 years')
plt.bar(x , pop.loc[:10, 'pop05'] / 1000, width, label = '2005 years')
plt.bar(x + width, pop.loc[:10, 'pop00'] / 1000, width, label = '2000 years')

plt.title("Population of major cities(2010)")
plt.xlabel('Municipality')
plt.ylabel('Population(in thousands)')
plt.xticks(x, labels = pop.loc[:10, 'municipality'])
plt.show()
```

```python
file = os.listdir()[-2]
outsource = pd.read_csv(file)
```

```python
plt.figure(figsize = (12, 8))
plt.plot(outsource['outsourcing'], outsource['delays'], 'o')
plt.show()
```

```python
plt.figure(figsize = (12, 8))
plt.plot(outsource['outsourcing'], outsource['delays'], 'o', color = 'r')
plt.xlabel('Maintenance outsourcing(%)')
plt.ylabel('Flight delays(%)')
plt.title('Outsourcing vs Flight delays')
plt.show()
```

```python
plt.figure(figsize = (12, 8))
plt.plot(outsource['outsourcing'], outsource['delays'], 'o', color = 'r')
plt.axhline(20, linestyle = '--', linewidth = 2, color = 'r')
plt.xlabel('Maintenance outsourcing(%)')
plt.ylabel('Flight delays(%)')
plt.title('Outsourcing vs Flight delays')
plt.show()
```
